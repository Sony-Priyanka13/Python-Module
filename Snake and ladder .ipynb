```python
# snake and ladder game
import random

def new_game():
    '''There are 2 players in game and one who reach 100 fast they are winners.To start game one must roll a dice of 6'''
    started1=False
    started2=False
    value1=0
    value2=0
    win1=False
    win2=False
    snakes={16:6,46:25,49:11,62:19,64:60,74:53,92:88,95:75,99:80}
    ladders={2:38,7:14,8:31,15:26,21:42,28:84,36:44,51:67,71:91,78:98,87:94}
    while True:
        if  not(started1):
            print("player1 is at position {}".format(value1))
            dice = random.randint(1,6)
            print("player1 rolled {}".format(dice))
            if dice==6:
                print("player1 stated the game")
                started1=True
                value1=1
        elif started1:   
            dice=random.randint(1,6)
            print("player1 is at position {}".format(value1))
            print("player1 rolled {}".format(dice))
            value1 +=dice
            if value1 in snakes:
                value1 = snakes[value1]
            if value1 in ladders:
                value1 = ladders[value1]
        print("player1 is at position {}".format(value1))
        if(value1>100):
            value1-=dice
        if(value1==100):
            print("player1 wins")
            win1=True
        if  win1:
            break
        if  not(started2):
            print("player2 is at position {}".format(value2))
            dice = random.randint(1,6)
            print("player2 rolled {}".format(dice))
            if dice==6:
                print("player2 stated the game")
                started2=True
                value2=1
        elif started2:
            print("player2 is at position {}".format(value2))
            dice=random.randint(1,6)
            print("player2 rolled {}".format(dice))
            value2 +=dice
            if value2 in snakes:
                value2 = snakes[value2]
            if value2 in ladders:
                value2 = ladders[value2]
        print("player2 is at position {}".format(value2))
        if(value2>100):
            value2-=dice
        if(value2==100):
            print("player2 wins")
            win2=True
        print("------------------")
        if  win2:
            break
        
            
```

```python
new_game()
```
